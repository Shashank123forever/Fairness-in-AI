#### loading packgaes

```python
from GoogleNews import GoogleNews
from Dbias.text_debiasing import *
```

#### functions

```python
def news(topic):
    googlenews = GoogleNews()
    googlenews = GoogleNews(lang='en', region='US')
    googlenews = GoogleNews(period='7d')
    googlenews.get_news(topic)
    result=googlenews.result()
    return result

def debiasing(x):
    suggestions = run(x)
    if suggestions == None:
        return ""
    else:
      all_suggestions = []
      for sent in suggestions[0:3]:
        all_suggestions.append(sent['Sentence'])
      return "\n\n".join(all_suggestions)
        
```

#### implementation

```python
# reading the latest news, enter the topic required
latest_news = news(topic = "Covid Omicron")
news_df = pd.DataFrame(latest_news)
news_df.head()
```

```python
# running Dbias to recommend unbiased/reduced bias news fragment
news_df['suggestion'] = news_df['title'].apply(lambda x: debiasing(x))
news_df[['title', 'suggestion']]
```

```python
# saving the dataframe to csv
news_df.to_csv("news-unbiased.csv")
```
